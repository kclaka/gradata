```python
import matplotlib.pyplot as plt
```

```python
# Open the data file
with open('Drop.log', 'r') as file:
    lines = file.readlines()

# Initialize empty lists to store times and packet counts
times = []
packets = []

# Parse the data
for line in lines:
    split_line = line.split(',')
    times.append(float(split_line[0].split(' ')[1][:-1]))  # remove 's' from time and convert to float
    packets.append(int(split_line[1].split(' ')[3]))  # get packet number and convert to int

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(times, packets, marker='o', linestyle='-', color='b')

# Set labels and title
plt.xlabel('Time (s)')
plt.ylabel('Dropped Packets')
plt.title('Dropped Packets over Time')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

def parse_file(file_name):
    # Initialize empty lists to store times and queue sizes
    times = []
    queue_sizes = []

    # Open the data file
    with open(file_name, 'r') as file:
        lines = file.readlines()

    # Parse the data
    for line in lines:
        split_line = line.split(',')
        times.append(int(split_line[0].split(' ')[1][:-1]))  # remove 's' from time and convert to int
        queue_sizes.append(int(split_line[1].split(' ')[3]))  # get queue size and convert to int

    return times, queue_sizes


# Parse the files
times_reno, queue_sizes_reno = parse_file('Queue.log')
times_cubic, queue_sizes_cubic = parse_file('quecubic.log')

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(times_reno, queue_sizes_reno, marker='o', linestyle='-', color='b', label='Reno')
plt.plot(times_cubic, queue_sizes_cubic, marker='o', linestyle='-', color='r', label='CUBIC')

# Set labels, title and legend
plt.xlabel('Time (s)')
plt.ylabel('Queue Size (packets)')
plt.title('Queue Size over Time')
plt.legend()

# Show the plot
plt.show()
```
